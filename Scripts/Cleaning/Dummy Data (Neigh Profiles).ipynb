## *Average income of neighbourhood*

```python
df_income = df_profiles[(df_profiles.Characteristic == 'Neighbourhood Number') | 
                       (df_profiles.Characteristic == 'Total income: Average amount ($)')]
```

```python
df_income = df_income.T.reset_index()
```

*Looking the head we exclude 1st few rows to select needed data*

```python
df_income.head(10)
```

```python
df_income = df_income.iloc[5:,:]
```

```python
df_income.columns = ['Neighborhood','Neighborhood Num','Average income']
```

```python
df_income.reset_index(inplace=True)
```

```python
df_income.drop(columns=['index'], axis = 1,inplace=True)
```

```python
df_income.head()
```

*Change the column types from string to integer*

```python
def strToInt(row):
    return int(row.replace(',',''))
```

```python
df_income['Average income'] = df_income['Average income'].apply(strToInt)
```

```python
df_income['Neighborhood Num'] = df_income['Neighborhood Num'].astype(int) 
```

```python
df_income.info()
```

```python
df_income.to_csv('../Data/Cleaned/Income.csv', index=False)
```
